```python
# ----------------------------------------------------
# 1) Import required libraries
# ----------------------------------------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 2) Create a sample dataset (you may replace with CSV)
# ----------------------------------------------------
np.random.seed(42)

X = np.random.rand(100, 5) * 10  # 100 rows, 5 features
df = pd.DataFrame(X, columns=[f"Feature_{i}" for i in range(1, 6)])

print("Sample Dataset:")
print(df.head())

# ----------------------------------------------------
# 3) Standardize the data (PCA requires scaling)
# ----------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# ----------------------------------------------------
# 4) Apply PCA (keep 2 principal components)
# ----------------------------------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# ----------------------------------------------------
# 5) Create a PCA dataframe
# ----------------------------------------------------
pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
print("\nPCA Results:")
print(pca_df.head())

# ----------------------------------------------------
# 6) Explained Variance
# ----------------------------------------------------
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

plt.figure(figsize=(7, 4))
plt.bar(["PC1", "PC2"], pca.explained_variance_ratio_)
plt.title("Explained Variance by Principal Components")
plt.ylabel("Variance Ratio")
plt.show()

# ----------------------------------------------------
# 7) Scatter plot of PCA components
# ----------------------------------------------------
plt.figure(figsize=(7, 5))
plt.scatter(pca_df["PC1"], pca_df["PC2"])
plt.title("PCA Scatter Plot (PC1 vs PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
